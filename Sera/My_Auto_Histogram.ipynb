```python
import numpy as np

from astropy.io import fits
import matplotlib.pyplot as plt
import statistics as pystats
```

```python
#File Name - 
def find_file(galaxy, band):
    dir = '/home/sera-mathew/Galaxies/raw_background_comps/'
    file_name = dir + galaxy + '_' + band + '_mjysr.fits'
    return file_name


w1 = fits.getdata(find_file('ic5332', 'w1'))
w1 = w1.flatten()
```

```python
# sort w1 and trim array
```

```python
minclip = -.01
maxclip = .05
bincount = 100
figx = 8
figy = 8

#calculate mean and median
mean = pystats.mean(w1)
median = pystats.median(w1)

# HELP what does this do? 
tfw1 = plt.figure(figsize=(figx, figy))
bins=np.linspace(minclip, maxclip, bincount)
Y_w1, BE_w1 = np.histogram(w1, bins=bins)
X_w1 = BE_w1[:-1]
plt.plot(X_w1, Y_w1, color = 'blue')

#plot mean and median
plt.axvline(mean, color = 'green', linestyle = 'dashed')
plt.axvline(median, color = 'black', linestyle = 'dashed')

plt.xlabel('')
plt.ylabel('')
plt.title('ic5332_w1')
plt.legend(['w1', 'mean', 'median'])
plt.show()
```
