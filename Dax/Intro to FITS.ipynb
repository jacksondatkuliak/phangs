# What is a FITS File?

A FITS (Flexible Image Transport System) file is a digital file format used primarily in astronomy for storing, transmitting, and manipulating scientific data. It is useful for images, spectral data, and other multidimensional data arrays, but it is also flexible enough to store a wide variety of data types.


FITS is an open standard maintained by the International Astronomical Union (IAU). This ensures compatibility and consistency across various platforms.


A FITS file consists of one or more Header/Data Units (HDUs). Each HDU contains a header followed by data:

- Header: The header is a series of 80-character ASCII card images, each containing a keyword, a value, and an optional comment. The header provides metadata about the data, such as size, dimensions, data type, and observational parameters.
- Data: The data can be an image (2D array), a table, or more complex data structures. The data is stored in binary format immediately following the header.


The FITS format supports a variety of data types and structures, making it highly versatile for different kinds of scientific data beyond just images. This includes time series, spectra, multi-dimensional data cubes, and more.


FITS files are self-describing, meaning that the headers contain all the necessary information to interpret the data correctly. This includes details such as the dimensions of the data array, scaling factors, and observational parameters.


Additional HDUs can be appended to a FITS file, allowing for the inclusion of extra data and metadata. This makes it possible to store related datasets within a single FITS file.

Below is an example of a FITS HDU of IC 5332, a spiral galaxy over 29 million light-years from Earth.

```python
import astropy.io.fits as fits
import matplotlib.pyplot as plt

# Open the FITS file
fits_file = 'ic5332_w2_mjysr.fits'  # Replace with your FITS file path
hdul = fits.open(fits_file)

# Print information about the FITS file
hdul.info()

# Access the primary HDU (Header/Data Unit)
primary_hdu = hdul[0]

# Print the header of the primary HDU
print(repr(primary_hdu.header))

# Access the data in the primary HDU
data = primary_hdu.data

# Print the shape and type of the data
print(f'Data shape: {data.shape}')
print(f'Data type: {data.dtype}')

# Close the FITS file
hdul.close()
```

Here is a breakdown of what each row means:

### Header Information Overview:


- Filename: ic5332_w2_mjysr.fits indicates the name of the FITS file
- No.: The index number of the HDU (Header/Data Unit). Here, 0 indicates the first (and primary) HDU.
- Name: The name of the HDU. PRIMARY indicates the primary HDU.
- Ver: Version number of the HDU. Here, it is 1.
- Type: Type of the HDU. PrimaryHDU indicates it is the primary data unit.
- Cards: Number of header cards (key-value pairs) in this HDU. Here, it is 42.
- Dimensions: The shape of the data array in this HDU, which is (655, 655).
- Format: Data type of the array elements. float64 indicates 64-bit floating point numbers.


### Primary Header


- SIMPLE: Indicates that the file conforms to the FITS standard. T stands for True.
- BITPIX: Specifies the data type of the image. -64 means the data is stored as 64-bit floating point numbers (IEEE double precision).
- NAXIS: Number of data axes. 2 indicates a two-dimensional array.
- NAXIS1: Length of the first data axis (655 pixels).
- NAXIS2: Length of the second data axis (655 pixels).
- EXTEND: Indicates that the FITS file may contain extensions beyond the primary HDU. T stands for True.
- COMMENT: Provides additional information or comments. Here, it references the definition of the FITS format.


### World Coordinate System (WCS) Information


- CTYPE1: Type of the first coordinate (Right Ascension, tangent plane projection).
- CTYPE2: Type of the second coordinate (Declination, tangent plane projection).
- CRVAL1: This represents the celestial longitude in degrees of the reference point for the image.
- CRVAL2: This represents the celestial latitude in degrees of the reference point for the image.
- CRPIX1: This indicates the pixel position along the horizontal axis where the reference point is located.
- CRPIX2: This indicates the pixel position along the vertical axis where the reference point is located.
- CD1_1: This coefficient describes the linear transformation applied to convert pixel coordinates to celestial coordinates.
- CD1_2:  This coefficient represents the second term of the linear transformation along the X-axis.
- CD2_1: This coefficient represents the first term of the linear transformation along the Y-axis.
- CD2_2: This coefficient describes the linear transformation along the Y-axis.
- IMAGEW: This indicates the width of the image in pixels.
- IMAGEH: This indicates the height of the image in pixels.
- WCSAXES:Indicates the number of dimensions in the WCS transformation.
- EQUINOX: The equinox is the reference point for measuring celestial coordinates.
- LONPOLE: This parameter defines the native longitude of the celestial pole in the coordinate system used for the image.
- LATPOLE: This parameter defines the native latitude of the celestial pole in the coordinate system used for the image.
- MAGZP: Specifies the magnitude zeropoint used for photometric calibration.
- UNW_SKY: Represents the background level subtracted from the coadded image, typically to remove sky background or instrumental noise.
- UNW_VER: Indicates the git revision or version of the unWISE code used to process the data. This helps track the software version for reproducibility.
- UNW_URL: Provides the URL to the repository containing the unWISE code.
- UNW_DVER: Specifies the version of the data model used by the unWISE code.
- UNW_DATE: Indicates the date and time when the unWISE processing was executed.
- UNW_FRO: Specifies the start index or identifier for the unWISE frames used in processing.
- UNW_FRN: Indicates the total number of unWISE frames included in the processing.
- UNW_MEDF: Specifies the size of the median filter used in the unWISE processing pipeline.
- UNW_BGMA: Indicates whether background matching was performed during the unWISE processing.
- REFEREN1 & REFEREN2:References the directory containing reference coadded images.
- EPOCH: Represents the epoch number associated with the dataset. The epoch is a specific moment in time, often used in astronomical observations to account for celestial motion.
- MJDMIN: Specifies the minimum Modified Julian Date (MJD) among the L1b frames included in the dataset
- MJDMAX: Specifies the maximum Modified Julian Date (MJD) among the L1b frames included in the dataset
- BAND: Specifies the WISE (Wide-field Infrared Survey Explorer) band number associated with the data. WISE observes the sky in multiple infrared bands, and this parameter identifies the specific band used.
- NMTOMJY: Provides the conversion factor from nanomaggies (a unit used in WISE data) to Megajanskys per steradian (MJy/sr), a common unit for infrared flux density.
- BUNIT: Specifies the units of the data, indicating that the values represent flux density in Megajanskys per steradian (MJy/sr).
- Data shape: This row indicates the shape or dimensions of the data array stored in the FITS file.
- Data type: This row specifies the data type or format of the values stored in the data array. The data type >f8 indicates that the values in the data array are stored as 64-bit floating-point numbers (float64 in Python). The > symbol signifies big-endian byte order, and f8 indicates a floating-point format with 8 bytes per element.

## Plotting FITS files

There are different scales that we can use to represent the data depending on what we want to look at. Below there are several examples.

### 1. Linear Scaling:
Linear scaling is the simplest and most common method for visualizing data. It preserves the original relationship between pixel values and intensities in the image. In linear scaling, each pixel value is directly mapped to a color intensity. The grayscale colormap is often used for linearly scaled images. Linear scaling is suitable for images with a wide dynamic range and evenly distributed pixel values.

### 2. Logarithmic Scaling:
Logarithmic scaling is employed when the data exhibits a wide range of values with significant variation. It compresses the dynamic range of the data, emphasizing details in the lower intensity regions while preventing saturation in the higher intensity regions. Logarithmic scaling is particularly effective for visualizing astronomical data, such as FITS images, where faint features need to be enhanced without losing information in brighter regions.

In logarithmic scaling:
- Low pixel values are stretched to occupy a larger range of intensities, enhancing visibility of faint features.
- High pixel values are compressed, reducing their impact on the overall visualization.
- Logarithmic scaling is often used in conjunction with linear scaling to enhance contrast in both dim and bright regions of an image.

### 3. Histogram Equalization:
Histogram equalization is a technique used to enhance the contrast of an image by redistributing pixel values in the histogram. It aims to make the cumulative distribution function of pixel intensities as uniform as possible. This process effectively spreads out the intensity values across the entire dynamic range, resulting in an image with improved contrast and detail.

In histogram equalization:
- The histogram of pixel intensities is computed.
- The cumulative distribution function (CDF) of the histogram is computed.
- Pixel values are mapped to new values based on their corresponding CDF values, effectively redistributing intensities to achieve a more uniform histogram.
- Histogram equalization is particularly useful for enhancing the visibility of details in images with low contrast or uneven lighting conditions.

Each scaling method offers advantages depending on the characteristics of the data and the visualization goals. It's essential to choose the most appropriate scaling technique to effectively convey the information contained within the image.

```python
import astropy.io.fits as fits
import matplotlib.pyplot as plt

# Open the FITS file
hdul = fits.open('ic5332_w2_mjysr.fits')
data = hdul[0].data

# Plot with linear scaling
plt.figure(figsize=(8, 8))
plt.imshow(data, cmap='gray')
plt.colorbar()
plt.title('Linear Scaling')
plt.show()

# Close the FITS file
hdul.close()
```

```python
import numpy as np

# Open the FITS file
hdul = fits.open('ic5332_w2_mjysr.fits')
data = hdul[0].data

# Apply logarithmic scaling to the data
data_log = np.log(data + 1)  # Adding 1 to avoid log(0)

# Plot with logarithmic scaling
plt.figure(figsize=(8, 8))
plt.imshow(data_log, cmap='gray')
plt.colorbar()
plt.title('Logarithmic Scaling')
plt.show()

# Close the FITS file
hdul.close()
```

```python
from skimage import exposure

# Open the FITS file
hdul = fits.open('ic5332_w2_mjysr.fits')
data = hdul[0].data

# Apply histogram equalization to the data
data_eq = exposure.equalize_hist(data)

# Plot with histogram equalization
plt.figure(figsize=(8, 8))
plt.imshow(data_eq, cmap='gray')
plt.colorbar()
plt.title('Histogram Equalization')
plt.show()

# Close the FITS file
hdul.close()
```

## Histograms


Histograms are helpful for visualizing the distribution of data by displaying the frequency or count of values within predefined bins. They provide insights into the central tendency, variability, and shape of the data distribution. Histograms enable quick identification of patterns, outliers, and skewness in the data, aiding in data exploration, analysis, and decision-making. Additionally, histograms can reveal underlying distributions, facilitating model selection and parameter estimation in statistical analysis.

```python
from astropy.io import fits
import matplotlib.pyplot as plt
import glob

# Open the selected FITS file
hdul = fits.open('ic5332_w2_mjysr.fits')
data = hdul[0].data

# Create a figure for the histogram
fig, ax = plt.subplots(figsize=(8, 6))

# Plot the histogram
ax.hist(data.flatten(), bins='auto', histtype='step', color='blue', lw=2, density=True)
ax.set_xlabel('Intensity')
ax.set_ylabel('Density')

# Set x-axis limits
ax.set_xlim(-10, 10)

# Plot mean and median as vertical lines
mean_val = data.flatten().mean()
median_val = np.median(data.flatten())
ax.axvline(x=mean_val, color='red', linestyle='--', label='Mean')
ax.axvline(x=median_val, color='green', linestyle='--', label='Median')
ax.legend()

plt.tight_layout()
plt.show()

# Close the FITS file
hdul.close()
```

As we can see, most of the data is centered around 0. This is the background of our FITS file and we will have to zoom in to get a better view of the background data.

```python
from astropy.io import fits
import matplotlib.pyplot as plt
import glob

# Open the selected FITS file
hdul = fits.open('ic5332_w2_mjysr.fits')
data = hdul[0].data

# Create a figure for the histogram
fig, ax = plt.subplots(figsize=(8, 6))

# Plot the histogram
ax.hist(data.flatten(), bins='auto', histtype='step', color='blue', lw=2, density=True)
ax.set_xlabel('Intensity')
ax.set_ylabel('Density')

# Set x-axis limits
ax.set_xlim(-.1, .1)

# Plot mean and median as vertical lines
mean_val = data.flatten().mean()
median_val = np.median(data.flatten())
ax.axvline(x=mean_val, color='red', linestyle='--', label='Mean')
ax.axvline(x=median_val, color='green', linestyle='--', label='Median')
ax.legend()

plt.tight_layout()
plt.show()

# Close the FITS file
hdul.close()
```

### Comparing W1 with W2 and W3 with W4


WISE (Wide-field Infrared Survey Explorer) captures images in multiple infrared bands, including WISE1 and WISE2. Each band captures different wavelengths of light, revealing different features and properties of celestial objects. Comparing images from different bands can provide a more comprehensive understanding of the object's physical characteristics, such as temperature, composition, and structure.


Different infrared bands can highlight different features of the same astronomical object. For instance, one band might show more details in the dust distribution, while another might be better for identifying star-forming regions. Graphical comparison helps in identifying and studying these features across different wavelengths.


Comparing WISE1 and WISE2 images can help in calibrating the data, ensuring that the measurements are consistent across different bands. It also helps in validating the presence of detected objects, as real astronomical objects should appear in multiple bands, whereas noise or artifacts might not.

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits

# Open the first FITS file
with fits.open('ic5332_w1_mjysr.fits') as hdul1:
    data1 = hdul1[0].data

# Open the second FITS file
with fits.open('ic5332_w2_mjysr.fits') as hdul2:
    data2 = hdul2[0].data

# Flatten the 2D arrays to 1D vectors
data1_flat = data1.flatten()
data2_flat = data2.flatten()

# Perform linear regression
slope, intercept = np.polyfit(data1_flat, data2_flat, 1)

# Plot the interpolated values against each other
plt.scatter(data1_flat, data2_flat, alpha=0.5)
plt.plot(data1_flat, slope * data1_flat + intercept, color='red')  # Adding the regression line
plt.xlabel('Clipped Values - ic5332_w1_mjysr.fits')
plt.ylabel('Clipped Values - ic5332_w2_mjysr.fits')
plt.title('Clipped Values Comparison with Linear Regression')
plt.grid(True)
plt.show()

# Print the equation of the best fit line
print(f"Best fit line equation: y = {slope:.2f}x + {intercept:.2f}")
```

Despite the variations at higher values, which is due to the different wavelengths, we can see that there is a linear fit that describes the relationship between the two files which means that the measurements across wavelengths are consistent
